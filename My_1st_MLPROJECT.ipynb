```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
dataset =pd.read_csv('Sale.csv')
df = pd.DataFrame(dataset)
df.head()
df_new = df.drop(columns =['Order Date','Order ID', 'Ship Date', 'Item Type', 'Sales Channel', 'Order Priority',
       'Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue','Region'])
```

```python
print(format(df.keys()))
print(format(df_new.keys()))
```

```python
df_new
```

```python
X =df_new.drop(columns =['Country'])
```

```python
y = df_new['Country']
print(y)
```

```python
model = DecisionTreeClassifier()
model.fit(X,y, )
```

```python
prediction = model.predict([[222243,90]])
```

```python
prediction
```

```python
print("ACCURACY:  {:.2f}".format(model.score(X,y)))
```
